```python
# ===============================
# 1. Importar librerías
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")

# ===============================
# 2. Cargar datos
# ===============================
# Cargamos los datos directamente desde la hoja de calculo de google
# URL de origen: https://docs.google.com/spreadsheets/d/1P0rPmzSC2GfbWvjCNo9a19DIr8tvbYPvOf9-nsoi6SM/edit?resourcekey=&gid=1249336249#gid=1249336249
# Se elimina /edit?usp=sharing
# Extraemos el id: 1P0rPmzSC2GfbWvjCNo9a19DIr8tvbYPvOf9-nsoi6SM
# definimos la variable sheet_id
sheet_id = '1P0rPmzSC2GfbWvjCNo9a19DIr8tvbYPvOf9-nsoi6SM'
# Leemos el df desde la url> https://docs.google.com/spreadsheets/d/ agregando este formato al final: [sheet_id}/export?format=csv
df = pd.read_csv(f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv")
```

```python
# ===============================
# 3. Exploración inicial
# ===============================
print("Dimensiones del dataset:", df.shape)
print("\nResumen de variables:")
print(df.info())
print("\nValores nulos por variable:")
print(df.isnull().sum())
```

```python
# ===============================
# 4. Limpieza básica
# ===============================

# Renombramos las columnas para que sean más fáciles de usar:
df = df.rename(columns={
    "1️⃣ Plataforma favorita": "plataforma",
    "2️⃣ Formato preferido": "formato",
    "3️⃣ Dispositivo principal": "dispositivo",
    "4️⃣ Horas al día": "horas_dia",
    "5️⃣ Tipo de acceso": "suscripcion",
    "6️⃣ Maratón o episodio diario": "estilo_consumo",
    "Edad": "edad",
    "Género": "genero"
})

# Estandarizamos columnas Categóricas

# Plataforma: Lowercase, strip y correcciones específicas
df['plataforma'] = df['plataforma'].str.lower().str.strip()
df['plataforma'] = df['plataforma'].replace({
    'youtube ': 'youtube',
    'magic': 'magis tv',
    'magic ': 'magis tv',
    'maggi': 'magis tv',
    'magic tv': 'magis tv',
})

# Definimos una paleta de colores personalizada para la identificación de plataformas
colores_plataformas = {
    'netflix': '#E50914',      # Rojo vibrante de Netflix
    'disney+': '#113CCF',      # Azul oscuro de Disney+
    'prime video': '#00A8E1',  # Turquesa/Azul de Prime Video (antes amarillo, ahora más azul)
    'hbo max': '#92278F',      # Púrpura/Violeta de Max/HBO Max
    'star+': '#C5C5C5',        # Gris oscuro/plata de Star+
    'paramount+': '#0064FF',   # Azul brillante de Paramount+
    'youtube': '#FF0000',      # Rojo de YouTube
    'apple tv+': '#1E1E1E',    # Negro/Gris oscuro de Apple TV+
    'magis tv': '#FFA500',      # Asumiremos un verde para este caso (ajustar si es necesario)
    'sensa': '#008000',
    'twitch': '#6441A5',
    'stremio': '#664181',
    'otro': '#808080',
    'no tengo': '#808081'
    # Agregar más plataformas y colores según sea necesarioda
}

# Formato: Strip y correcciones
df['formato'] = df['formato'].str.strip()
df['formato'] = df['formato'].replace({
    'Anime': 'Serie (varias temporadas)',
    'Música': 'otro_formato',
})

# Eliminamos la fila de "Música" (otro_formato)
df = df[df['formato'] != 'otro_formato']

# Dispositivo: Lowercase y strip
df['dispositivo'] = df['dispositivo'].str.lower().str.strip()

# Género y Edad: Reemplazamos "Prefiero no responder" por NaN y eliminar esas filas
#df['genero'] = df['genero'].replace('Prefiero no responder', np.nan)
#df['edad'] = df['edad'].replace('Prefiero no responder', np.nan)
#df.dropna(subset=['genero', 'edad'], inplace=True)

# Estandarización de Horas_dia
# Convertiremos los rangos de horas a un orden lógico para gráficos.
# Definimos el orden deseado para la columna de horas
orden_horas = ['Menos de 1 hora', '1–2 horas', '3–4 horas', '5-6 horas','Más de 6 horas']

# Convertir la columna a un tipo de dato 'Category' con el orden definido
df['horas_dia'] = pd.Categorical(df['horas_dia'], categories=orden_horas, ordered=True)

# Estandarización de Edad
# Hacemos lo mismo con la edad para asegurar un orden correcto en las visualizaciones
# Definir el orden deseado para la columna de edad
orden_edad = ['-15 años','15–24 años', '25–34 años', '35–44 años', '45–54 años', '55 años o más']

# Convertir la columna a un tipo de dato 'Category' con el orden definido
df['edad'] = pd.Categorical(df['edad'], categories=orden_edad, ordered=True)
```

```python
# ===============================
# 5. Análisis descriptivo
# ===============================

# Conteo de respuestas por variable categórica
for col in ["plataforma", "formato", "dispositivo", "horas_dia", "suscripcion", "estilo_consumo", "edad", "genero"]:
    if col in df.columns:
        print(f"\nFrecuencia de {col}:")
        print(df[col].value_counts())
print('\n')
```

```python
#PROBANDO
# ===============================
# 6. Visualizaciones
# ===============================

import seaborn as sns
import matplotlib.pyplot as plt

# Normalizar categorías (ejemplo: unificar "otros" y distinguir minúsculas/mayúsculas)
df["plataforma"] = df["plataforma"].str.strip().str.lower().replace({"otros": "Otros"})
df["formato"] = df["formato"].str.strip().str.lower().replace({"otros": "Otros"})
df["dispositivo"] = df["dispositivo"].str.strip().str.lower().replace({"otros": "Otros"})
df["estilo_consumo"] = df["estilo_consumo"].str.strip().str.lower().replace({"otros": "Otros"})

# Función auxiliar para agregar etiquetas de cantidad
def agregar_etiquetas(ax):
    for p in ax.patches:
        ax.annotate(
            int(p.get_height()),
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='bottom',
            fontsize=10, color='black', xytext=(0, 5),
            textcoords='offset points'
        )

# Plataforma favorita
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x="plataforma", order=df["plataforma"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Plataforma de streaming favorita")
ax.set_ylabel("Cantidad")  # cambiar count por cantidad
agregar_etiquetas(ax)
plt.show()
print('\n')

# Formato preferido
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x="formato", order=df["formato"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Formato preferido en streaming")
ax.set_ylabel("Cantidad")
agregar_etiquetas(ax)
plt.show()
print('\n')

# Dispositivo principal
plt.figure(figsize=(6,4))
ax = sns.countplot(data=df, x="dispositivo", order=df["dispositivo"].value_counts().index)
plt.xticks(rotation=45)
plt.title("Dispositivo más usado")
ax.set_ylabel("Cantidad")
agregar_etiquetas(ax)
plt.show()
print('\n')

# Estilo de consumo (maratón vs episodio)
plt.figure(figsize=(6,4))
ax = sns.countplot(data=df, x="estilo_consumo", order=df["estilo_consumo"].value_counts().index)
plt.title("Preferencia: Maratón vs Episodio diario")
plt.xticks(rotation=45)
ax.set_ylabel("Cantidad")
agregar_etiquetas(ax)

# Ajuste para evitar solapamiento
plt.tight_layout()
plt.show()
print('\n')
```

```python
# Análisis Exploratorio y Contraste de Hipótesis
# Utilizaremos gráficos de barras (countplot y barplot de Seaborn) y tablas de contingencia para explorar las relaciones entre las variables.

# Plataforma Más Popular (Contraste de Hipótesis: Plataformas más populares)
# Hipótesis: Netflix será la favorita, con fuerte presencia de Disney+ y Prime Video en grupos jóvenes.

# Filtramos solo los colores presentes en la muestra para la paleta
plataformas_en_data = df['plataforma'].unique()
colores_para_grafico = {
    p: c for p, c in colores_plataformas.items() if p in plataformas_en_data
}

plt.figure(figsize=(10, 6))
sns.countplot(
    y='plataforma',
    data=df,
    order=df['plataforma'].value_counts().index,
    # Usamos la solución al warning: asignamos 'plataforma' a hue
    hue='plataforma',
    #legend=False,
    palette=colores_para_grafico
)
plt.title('Plataforma de Streaming Favorita')
plt.xlabel('Número de Respuestas')
plt.ylabel('Plataforma')
plt.show()

print(f"La plataforma más popular según la muestra es: {df['plataforma'].mode()[0]}")
print(df['plataforma'].value_counts())
```

```python
# ===============================
# 7. Análisis cruzado
# ===============================

# Ejemplo: plataforma favorita según edad
if "edad" in df.columns:
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x="edad", hue="plataforma")
    plt.xticks(rotation=45)
    plt.title("Plataforma favorita según rango etario")
    plt.show()
    print('\n')

# Ejemplo: formato preferido según horas al día
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="horas_dia", hue="formato")
plt.title("Formato preferido según horas de consumo diario")
plt.show()
print('\n')
```

```python
plt.figure(figsize=(14, 7)) # Aumentamos el ancho para acomodar las etiquetas del eje X

# Gráfico de Barras Agrupadas Verticales:
# X: La variable a contar (Plataforma)
# Hue: La variable de segmentación (Edad)
sns.countplot(
    x='plataforma', # ¡Cambiado de 'y' a 'x'!
    hue='edad',
    data=df,
    # Mantenemos el orden de las plataformas en el eje X
    order=df['plataforma'].value_counts().index,
    palette='Spectral'
)

plt.title('Plataforma Favorita Segmentada por Grupo de Edad', fontsize=16)
plt.ylabel('Número de Respuestas', fontsize=12) # El conteo va en Y
plt.xlabel('Plataforma', fontsize=12)         # La categoría va en X
plt.legend(title='Edad', loc='upper right')

# Giramos las etiquetas del eje X para evitar que se superpongan
plt.xticks(rotation=45, ha='right')

plt.tight_layout() # Asegura que las etiquetas rotadas se muestren completamente
plt.show()
```

```python
# --- Asumiendo que el DataFrame 'df' tiene ahora la columna 'edad' correctamente categorizada ---

# 1. Crear la tabla de conteo cruzado (Crosstab)
conteo_plataforma_edad = pd.crosstab(df['plataforma'], df['edad'])

# 2. Ordenar las filas por el total de respuestas
conteo_plataforma_edad['Total'] = conteo_plataforma_edad.sum(axis=1)
conteo_plataforma_edad = conteo_plataforma_edad.sort_values(by='Total', ascending=False).drop(columns='Total')

# 3. Generar el Gráfico de Barras Apiladas
plt.figure(figsize=(12, 8))

ax = conteo_plataforma_edad.plot(
    kind='barh',
    stacked=True,
    colormap='Spectral',
    ax=plt.gca()
)

plt.title('Distribución de Plataforma Favorita Apilada por Edad (Completo)', fontsize=16)
plt.xlabel('Número Total de Respuestas', fontsize=12)
plt.ylabel('Plataforma', fontsize=12)
plt.legend(title='Edad', bbox_to_anchor=(1.05, 1), loc='upper left')

# Agregar el total de conteo al final de cada barra
for i, total in enumerate(conteo_plataforma_edad.sum(axis=1)):
    ax.text(
        total + 1,
        i,
        f'{total}',
        va='center',
        fontsize=10,
        fontweight='bold'
    )

plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```

```python
# 1. Filtrar los géneros principales para un gráfico más limpio
df_filtered = df[df['genero'].isin(['Femenino', 'Masculino'])]

# 2. Crear el Countplot (Gráfico de Frecuencias) con el 'hue' (color/separador)
plt.figure(figsize=(12, 7))

# Invertir la paleta de colores
palette_invertida = sns.color_palette('Pastel1', n_colors=2)[::-1]

sns.countplot(
    x='formato',
    hue='genero',
    data=df_filtered,
    palette=palette_invertida,
    order=df_filtered['formato'].value_counts().index
)

plt.title('Formato de Contenido Preferido por Género', fontsize=18)
plt.xlabel('Formato de Contenido', fontsize=14)
plt.ylabel('Frecuencia (Número de Encuestados)', fontsize=14)
plt.xticks(rotation=30, ha='right')
plt.legend(title='Género', loc='upper right')

# Añadir etiquetas de valor
ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()

# Análisis de Resultados
formato_genero_crosstab = pd.crosstab(df_filtered['formato'], df_filtered['genero'], normalize='columns') * 100
print("\n--- Porcentaje de Preferencia de Formato por Género ---")
print(formato_genero_crosstab.round(1))
```
